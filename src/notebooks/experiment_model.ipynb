```python
from faster_whisper import WhisperModel

model_size = "small"

model = WhisperModel(model_size, device="cuda", compute_type="float16")


segments, info = model.transcribe(r"dataset/sample.wav", word_timestamps=True)

for segment in segments:
    print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text)) 
    for word in segment.words:
        print("[%.2fs -> %.2fs] %s" % (word.start, word.end, word.word))
```

```python
from pydub import AudioSegment, silence

myaudio = AudioSegment.from_wav("dataset/sample.wav") 

dBFS = myaudio.dBFS
silent_segments = silence.detect_silence(myaudio, min_silence_len=1000, silence_thresh=dBFS-16)

silent_segments = [((start/1000),(stop/1000)) for start,stop in silent_segments]

print(f"Detected silent segments (in seconds): {silent_segments}")
```

```python
non_silent_segments = []

for i in range(len(silent_segments) - 1):
    end_current = silent_segments[i][1]
    start_next = silent_segments[i+1][0]
    non_silent_segments.append((end_current, start_next))

print(non_silent_segments)
```
